```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("2022_forbes_billionaires.csv",usecols=["rank","name","networth","age","country","source","industry"])
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().any()
```

```python
df["networth"] = df["networth"].astype("string").str.replace(r"[$B]","",regex=True).astype("float")
```

```python
df.sort_values("networth",ascending=False)
```

# Описательная статистика

```python
# Количество миллиардеров в списке
tottal_billionaires = len(df)
print('Кол-во миллиардеров =',tottal_billionaires)
```

```python
#Средний, медианный возраст миллиардера
mean_age = df["age"].mean()
median_age =df["age"].median()
print('Средний возраст =',mean_age)
print('Медиана =',median_age)
```

```python
#Среднее, минимальное и максимальное состояние
mean_networth = df["networth"].mean()
max_networth = df["networth"].max()
min_networth = df["networth"].min()

print('Среднее =',mean_networth)
print('Максимальное =',max_networth)
print('Минимальное =',min_networth)
```

```python
#Топ-10 по состоянию
df.head(10)
```

# География миллиардов

```python
# Топ 5 стран по количесвтву миллиардеров
df.groupby("country")["name"].count().sort_values(ascending=False).head(5)
```

```python
# Топ стран по суммарному состоянию
df.groupby("country")["networth"].sum().sort_values(ascending=False).head(5)
```

```python
# ВИзуализация итогов pie chart
top_country = df.groupby("country")["networth"].sum().sort_values(ascending=False).head(5)

plt.pie(top_country, labels=top_country.index, autopct='%1.1f%%', startangle=140)
plt.title("Топ 5 стран")
```

```python
# Визуализация итогов bar chart
top5 = df[["networth","name"]].head(5)
plt.bar(top5["name"],top5['networth'],color='skyblue')
plt.title("Топ 5 миллиардеров")
plt.xlabel("миллиардеры")
plt.ylabel("сотояние млрд.")
plt.xticks(rotation=45)
```

# Связь возраста и состояния

```python
plt.figure(figsize=(8,7))
plt.scatter(x=df["networth"],y=df["age"], color="skyblue")
plt.xlabel("Состояние")
plt.ylabel("Возраст")
plt.title("Зависимость возраста и состояния")
```

```python
df_corr = df[["networth","age"]].corr()
df_corr
```

```python
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x="networth", y="age", scatter_kws={"alpha": 0.5}, line_kws={"color": "red"})
plt.title("Линейная регрессия: возраст vs состояние")
plt.xlabel("Возраст")
plt.ylabel("Состояние (млрд $)")
plt.show()
```

# Источник состояния

```python
source_sum = df.groupby("source")["networth"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(7,6))
plt.bar(source_sum.index,source_sum.values)
plt.title("Источник состояния")
plt.xticks(rotation=45)
```

```python

# Распределение
bill_in_source = df.groupby("source")["name"].count().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
bars = plt.bar(bill_in_source.index, bill_in_source.values, color="skyblue")
plt.title("Кол-во миллиардеров по источникам")
plt.xlabel("Источник состояния")
plt.xticks(rotation=45, ha='right')

# Добавление значений над столбцами
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1,
             f'{int(height)}', ha='center', va='bottom', fontsize=10)

# Убираем левую ось и цифры на ней
ax = plt.gca()
ax.spines['left'].set_visible(False)       # убрать линию оси Y
ax.yaxis.set_visible(False)                # убрать метки и деления по Y
ax.spines['right'].set_visible(False)      # также убрать правую рамку

plt.tight_layout()
plt.show()


```

```python
industry_treemap = df.groupby("industry")["networth"].sum().sort_values(ascending=False).head(10)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

top_sources = df.groupby("source")["networth"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_sources.values, y=top_sources.index, palette=sns.color_palette("rocket", n_colors=10))
plt.title("Топ-10 источников состояния")
plt.xlabel("Суммарное состояние (млрд $)")
plt.ylabel("Источник")
plt.tight_layout()
plt.show()
```

#  Отчет по анализу миллиардеров Forbes 2022

1. Введение

В данном отчете представлен анализ списка миллиардеров Forbes 2022 года. Цель — выявить ключевые характеристики, закономерности и взаимосвязи, касающиеся возраста, состояния, стран и источников богатства.

2. Описательная статистика

Общее количество миллиардеров в выборке: около 2600 человек.

Средний возраст миллиардеров: ~60 лет.

Медианный возраст близок к среднему, что говорит о равномерном распределении возрастов.

Среднее состояние составляет порядка десятков миллиардов долларов.

Максимальное состояние (у лидера) превышает 200 млрд $.

Минимальное состояние около 1 млрд $.

3. География богатства

Топ-5 стран по количеству миллиардеров:

США

Китай

Германия

Индия

Россия

США также лидируют по суммарному состоянию миллиардеров, значительно опережая остальные страны.

Топ-5 стран аккумулируют основную часть мирового миллиардного капитала, что видно на круговой диаграмме.

4. Топ миллиардеров

Лидеры списка — Илон Маск, Джефф Безос и другие — имеют состояние, значительно превышающее большинство остальных миллиардеров.

Топ-5 миллиардеров заметно выделяются как по абсолютным значениям капитала, так и по влиянию на индустрии.

5. Взаимосвязь возраста и состояния

Корреляция между возрастом и состоянием близка к нулю (~0.04), что указывает на отсутствие значимой зависимости.

Это значит, что как молодые, так и пожилые миллиардеры могут иметь как высокое, так и низкое состояние.

Графики рассеяния и линейной регрессии показывают широкий разброс данных.

6. Источники состояния

Топ-10 источников состояния включают: технологии, инвестиции, производство, моду и ритейл.

Источники с наибольшим суммарным капиталом совпадают с теми, где сосредоточено больше всего миллиардеров.

Распределение состояния внутри источников часто неравномерно — несколько очень богатых лидеров и много менее состоятельных.

7. Индустрии

Анализ индустрий показывает, что ключевыми секторами являются: технологии, мода, инвестиции и производство.

Эти индустрии аккумулируют основную часть миллиардного капитала.

8. Общие выводы

Миллиардеры — это разнообразная по возрасту и индустриям группа, без явной зависимости возраста от уровня состояния.

Основные центры накопления богатства — США и Китай.

Главные драйверы капитала — технологический сектор и инвестиции.

Многообразие источников и индустрий свидетельствует о разнообразии путей к большому состоянию.
